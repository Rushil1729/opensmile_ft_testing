```python
import opensmile
import plotly.express as px
import plotly.graph_objects as go

# initialized smile object with feature set
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors
)

features = [  'Loudness_sma3',  'jitterLocal_sma3nz']

#y holds features of v1_01 to v5_01
y = []
for i in range(1,2): #Change back to 5 when required
    df = smile.process_file(f'./v{i}/v{i}_03.mp3')
    y.append(df)

DURATION = 0.02 #duration of each window
time_list = []
for j in range(y[0].shape[0]):
    time_list.append(f"{j/100}")

```

```python

#building figure
fig = go.Figure()
for x in features:
    fig.add_trace(
        go.Scatter(
            x = time_list,
            y = y[0][x],
            mode = 'lines',
            name = f'{x}'
        )
    )
fig.update_layout(
    height=500,
    width=1100,
)

fig.show()
```
