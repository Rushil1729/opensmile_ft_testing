```python
import opensmile
import plotly.express as px

smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors
)

list_of_features = ['Loudness_sma3']
v1feature = smile.process_file('./v1/v1_02.mp3')
v2feature = smile.process_file('./v2/v2_02.mp3')
v4feature = smile.process_file('./v4/v4_02.mp3')
v5feature = smile.process_file('./v5/v5_02.mp3')

v1feature.head()
```

```python

time_list = []
for i in range(v2feature.shape[0]):
    time_list.append(f"00:00:0{i/100}")

v2feature['Time'] = time_list
px.line(v2feature,y=list_of_features,x='Time',height=600,width=1000)
```

```python

time_list = []
for i in range(v1feature.shape[0]):
    time_list.append(f"00:00:0{i/100}")

v1feature['Time'] = time_list
px.line(v1feature,y=list_of_features,x='Time',height=600,width=1000)
```

```python

time_list = []
for i in range(v4feature.shape[0]):
    time_list.append(f"00:00:0{i/100}")

v4feature['Time'] = time_list
px.line(v4feature,y=list_of_features,x='Time',height=600,width=1000)
```

```python

time_list = []
for i in range(v5feature.shape[0]):
    time_list.append(f"00:00:0{i/100}")

v5feature['Time'] = time_list
px.line(v5feature,y=list_of_features,x='Time',height=600,width=1000)
```
